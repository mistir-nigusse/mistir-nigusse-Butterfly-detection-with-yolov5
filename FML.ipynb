````python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

folder30 = Path('runs/detect/ep30_preds/labels')
folder50 = Path('runs/detect/ep50_preds/labels')

det30 = len(list(folder30.glob('*.txt')))
det50 = len(list(folder50.glob('*.txt')))

total = 18
miss30 = total - det30
miss50 = total - det50

labels = ['Detected','Missed']
counts30 = [det30, miss30]
counts50 = [det50, miss50]
x = np.arange(len(labels))
w = 0.35

plt.figure()
plt.bar(x - w/2, counts30, w, label='exp_ep30')
plt.bar(x + w/2, counts50, w, label='exp_ep50')
plt.xticks(x, labels)
plt.ylabel('Number of Test Images')
plt.title('Per-Image Detection Outcomes Comparison')
plt.legend()
plt.show()
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

results_csv = Path('runs/train/exp_ep50/results.csv')
df = pd.read_csv(results_csv, skipinitialspace=True)

plt.figure()
plt.plot(df['epoch'], df['train/box_loss'], label='Box')
plt.plot(df['epoch'], df['train/obj_loss'], label='Obj')
plt.plot(df['epoch'], df['train/cls_loss'], label='Cls')
plt.title('Training Loss')
plt.legend()
plt.savefig('/content/loss_curves.png', bbox_inches='tight')
plt.show()

plt.figure()
plt.plot(df['epoch'], df['metrics/mAP_0.5'],      label='mAP@0.5')
plt.plot(df['epoch'], df['metrics/mAP_0.5:0.95'], label='mAP@0.5–0.95')
plt.title('Validation mAP')
plt.legend()
plt.savefig('/content/validation_map.png', bbox_inches='tight')
plt.show()
````
